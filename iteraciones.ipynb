## ☝ Iteraciones con ciclos For

En ocasiones necesitamos realizar la misma operación en múltiples ocasiones, en general, y podemos o no saber el número de veces que necesitamos realizarlo.

Vamos a crear una pequeña tabla para que nos ayude en nuestros ejemplos.

```
nombres <- c("Jorge", "Pedro", "Itzel", "Cristina","Manuel", "Leticia")
edad <-c(33, 13, 21, 18, 65, 15)

df <- data.frame(NOMBRE = nombres, EDAD = edad)

df
```


````r
nombres <- c("Jorge", "Pedro", "Itzel", "Cristina","Manuel", "Leticia")
edad <-c(33,13,21,18,65,15)

df <- data.frame(NOMBRE = nombres, EDAD = edad)

df
````

A continuación yo quisiera saber en que año nació cada una de las personas de la tabla para ello voy a utilizar el __ciclo for__ en dicho ciclo le voy a solicitar a la computadora que realice una o mas acciones un número determinado de veces.

Como convención se utiliza la variable arbitraria `i` como un contador que nos permite llevar la cuenta del numero de veces que ocurre la acción deseada.

```
# En la primera linea definimos el número de veces que se
# ejecutará el código dentro de los corchetes curvos `{ }`
# lo leemos cómo para cada i en el rango del 1 al número de filas
# que haya en el objeto `df` realizar lo siguiente:

for( i in 1:nrow(df){  

print(df$NOMBRE[i])

# Para este ejemplo la acción a realizar es
# imprimir el elemento "i" de la columna "NOMBRE",
# dónde "i" será el número 1 la primera vez que se ejecute el
#código, 2 la segunda, 3 la tercera y a así sucesivamente.

} # no olvides cerrar los corchetes


````r
for (i in 1:nrow(df)){
    print(df$NOMBRE[i])
}
````

Esto lo pudimos haber hecho simplemente usando `print(df$NOMBRE)` con un resultado similar.

Sin embargo los ciclos facilitan realizar operaciones mas complejas.

En este caso lo que quiero realizar es lo siguiente:

En caso de que la edad de alguien sea menor que 18 años, que me informe en pantalla el nombre seguido de la oración "es menor de edad"

```
for (i in 1:nrow(df)){
    if (df$EDAD[i] < 18 ){
      print(paste(df$NOMBRE[i], "es menor de edad"))
    }
}
```

````r
for (i in 1:nrow(df)){
    if (df$EDAD[i] < 18 ){
      print(paste(df$NOMBRE[i], "es menor de edad"))
    }
}
````

Yo puedo iniciar una columna vacia en un data frame o un tibble usando una asignación a un nombre de columna no existente el resultado del comando `c()` sin argumentos.

````r
class(df$EDAD[i])
````

````r
df$FECHA_NACIMIENTO <- c()

fecha_actual <- 2025

for(i in 1:nrow(df)){
  df$FECHA_NACIMIENTO[i] <- fecha_actual-df$EDAD[i]
}

df
````

## ☝ Iteraciones con While

`while` es un tipo de ciclo que funciona de manera similar a `for` con la excepción de que no es necesario saber cuántas veces se llevará a cabo una acción.

`while` ejecutará un código un número indeterminado de veces **hasta que una condición de paro se cumpla**

☢ **PRECAUCIÓN** : Es muy importante que verifiques que la condición de paro va a ocurrir, de lo contrario se va a ocasionar un [ciclo infinito](https://es.wikipedia.org/wiki/Bucle_infinito)

```
paro <- 1

while(paro != 10){
  paro <- sample(1:10, 1)
  print(paro)
}
```

En el código de arriba , el ciclo se ejecuta en tanto el valor de la variable `paro` sea distinto de 10, y el valor de dicha variable cambia de manera aleatoria tomando un número del 1 al 10 cada vez que se ejecuta el ciclo con `sample()`

Nota que definimos `paro` antes de iniciar el ciclo `while` y redefinimos su valor dentro del ciclo.

Definir `paro` antes es necesario para evitar el error de objeto no encontrado.

y redefinirlo es necesario para evitar un **bucle infinito**

````r
paro <- 1

while(paro != 10){
  paro <- sample(1:10, 1)
  print(paro)
}
````
