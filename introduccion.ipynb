## ☝ Operador de entubamiento (*Pipe*)

|Operador|Significado|
|:------:|---------|
|\|>| toma la salida de la operación de la izquierda y la utiliza como la entrada de la operación de la derecha |

**Nota:** Antes de R 4.1 no funciona, pero existe el operador `%>%` que únicamente sirve para determinadas funciones pero aún es muy frecuente encontrarlo en diferentes páginas y documentación, se debe preferir el mas nuevo `|>`


## 📖 Anidación vs entubamiento

En la anidación la función mas "adentro" se ejecuta primero:



```
tercer_función(segunda_funcion(primer_funcion(datos)))
```

En el entubamiento la ejecución es de izquierda a derecha



```
datos |> primer_funcion() |> segunda_funcion() |> tercer_funcion()
```
