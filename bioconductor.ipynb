## ☝ Bioconductor

[Link oficial](https://www.bioconductor.org)


*    Es una colección de paquetes destinados para el uso en el análisis de datos
biológicos.

*    Su software ha sido citado mas de 30,000 veces en pubmed

*    Instalador multisistema operativo



La función del instalador es necesario instalarla **una** vez pero siempre nos van a sugerir hacerlo en la documentación, esto es únicamente en caso de que no la tengamos instalada.

Ejemplos:

[DESeq2](https://bioconductor.org/packages/release/bioc/html/DESeq2.html)

[GSEABase](https://bioconductor.org/packages/devel/bioc/html/GSEABase.html)

[csaw](https://bioconductor.org/packages/release/bioc/html/csaw.html)

```r
if (!require("BiocManager", quietly = TRUE))
    install.packages("BiocManager")
BiocManager::install(version = "3.20")
```

Podemos ver los paquetes disponibles para instalar en la [página web](https://bioconductor.org/packages/release/BiocViews.html) o utilizando la función `available()`

**Nota** escribir `BiocManager::available()` nos permite usar la función `available()` del paquete `BiocManager` sin tener que cargar todas sus funciones usando `library()`
